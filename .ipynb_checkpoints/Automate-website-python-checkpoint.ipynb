```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome()
```

```python
# Set path website

#driver.get("https://shop.line.me/@442umjkq")
#driver.get("https://shop.line.me/@mcjeans_official/")
driver.get('https://shop.line.me/@611thbxn')
#driver.get("http://shop.line.me/@yuedpao/")
```

```python
# Login

# Define some var to used in login phease
email_used = "gustheworld6@gmail.com"
pwd_used = "@Zedbot130"

# Setup wait to wati login authen and current window 
wait = WebDriverWait(driver, 100)
actions = ActionChains(driver)      # action controll define
current_window = driver.current_url

# Refresh when popup show
while (driver.find_element(By.XPATH, "//div[@data-atd='share-modal-bottom-sheets']").is_displayed() or driver.find_element(By.XPATH, "//div[@data-atd='lm-greeting-bottom-sheets']").is_displayed() or driver.find_element(By.XPATH, "//div[@data-atd='send-gift-intro-modal-bottom-sheets']").is_displayed()):    
    time.sleep(0.5)
    driver.refresh()
    wait.until(EC.url_to_be(current_window))

# Find cart button to login
driver.find_element(By.XPATH, "//button[@data-atd='cart-button']").click()
driver.find_element(By.XPATH, "//button[@data-atd='modal-login-login-button']").click()

# Find login element
text_field_element = driver.find_elements(By.XPATH, "//fieldset[@class='MdFormGroup01']/div")

# Save input text
text_field_element[0].click()
actions.send_keys(email_used).perform()

text_field_element[1].click()
actions.send_keys(pwd_used).perform()

# Then click to login and authen with line applicaiton
text_field_element[2].find_element(By.XPATH, "//button").click()
```

```python
# Pre-pares to select product
wait.until(EC.url_to_be(current_window))

# wait for refresh popup
time.sleep(2.5)

while (driver.find_element(By.XPATH, "//div[@data-atd='share-modal-bottom-sheets']").is_displayed() or driver.find_element(By.XPATH, "//div[@data-atd='lm-greeting-bottom-sheets']").is_displayed() or driver.find_element(By.XPATH, "//div[@data-atd='send-gift-intro-modal-bottom-sheets']").is_displayed()):
    time.sleep(0.5)
    driver.refresh()
    wait.until(EC.url_to_be(current_window))
```

```python
# Information store

# Define some variable to use for keep product
specific_word_1 = "BUCKET HAT"
specific_word_2 = "Welcome to the jungle"

# OPTIONAL WORD ------------
#specific_word_3 = "Purple"
#specific_word_4 = "Deep sea"
# ----------------------------

# Array to keep product we want to select
real_link = []

# Find all product element in page
elements_case = driver.find_elements(By.XPATH,"//ul[@class='flex px-4 flex-wrap justify-between']/li")

# Check condition every product with specific word 
for element in elements_case:
    
    # WORD : 1
    if((specific_word_1 in element.text) and ("Sold Out" not in element.text)):
        real_link.append(element.find_element(By.TAG_NAME, "a"))
    
    # WORD : 2
    if((specific_word_2 in element.text) and ("Sold Out" not in element.text)):
        real_link.append(element.find_element(By.TAG_NAME, "a"))
    
print(len(real_link))
```

```python
# Access item with condition
# driver.switch_to.window(tab_handle[0])

actions = ActionChains(driver)
wait = WebDriverWait(driver, 10)

select_size = "L"

for element in real_link:
    
    # Create product link in new tab
    element.send_keys(Keys.CONTROL + Keys.RETURN)
    
    # Get current window handle
    tab_handle = driver.window_handles
    print(tab_handle)
    
    # Switch to another window 
    driver.switch_to.window(tab_handle[-1])

    # Add variant option
    add_to_cart = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@data-atd='add-to-cart-button']")))
    
    # Click many time to access cart item option
    while True:
        class_check = driver.find_element(By.XPATH, "//div[@data-atd='variant-modal-bottom-sheets']").get_attribute('class')
        
        if "pointer-events-none opacity-0 -z-10" not in class_check:
            break
        else:
            add_to_cart.click()
            time.sleep(0.3)
    
    # Search for all option in variant
    variant_option = driver.find_elements(By.XPATH, "//div[@id='modal-content']/div")
    
    time.sleep(0.3)
    
    # Check if doesn't have any text in it.
    while True:
        if variant_option[1].text != "":
            break
    
    # Loop for tick right option
    for option in variant_option:
    
        # OPTION : SIZE
        if "SIZE" in option.text:
            # Check finish
            check = False
            
            # Find option in size
            select_list = option.find_elements(By.XPATH, "./div[@class='flex flex-wrap']/button")
            
            # Search for right choice
            for choice in select_list:
                
                # Check right choice then access and break loop
                if "L" in choice.text and ("pointer-events-none" not in choice.get_attribute('class')):
                    choice.click()
                    check = True
                    break
            
            if not check:
                for any_choice in select_list:        
                    print(any_choice.text)
                    if "pointer-events-none" not in any_choice.get_attribute('class'):
                        any_choice.click()
                        break
        
        # OPTION : COLOR
        elif "COLOR" in option.text:
            print('color')
            
        # OPTION : 3
        elif "สี" in option.text:
            # Check finish
            check = False
            
            select_list = option.find_elements(By.XPATH, "./div[@class='flex flex-wrap']/button")
            for choice in select_list:
                if "ยีนส์" in choice.text and ("pointer-events-none" not in choice.get_attribute('class')):
                    choice.click()
                    check = True
                    break
            
            if not check:
                for any_choice in select_list:        
                    print(any_choice.text)
                    if "pointer-events-none" not in any_choice.get_attribute('class'):
                        any_choice.click()
                        break
        # OPTION : 4
        elif "ขนาด" in option.text:
            # Check finish
            check = False
            
            select_list = option.find_elements(By.XPATH, "./div[@class='flex flex-wrap']/button")
            for choice in select_list:
                if "31" in choice.text and ("pointer-events-none" not in choice.get_attribute('class')):
                    check = True
                    choice.click()
                    break
                    
            if not check:
                for any_choice in select_list:        
                    print(any_choice.text)
                    if "pointer-events-none" not in any_choice.get_attribute('class'):
                        any_choice.click()
                        break
                        
        # They not have any option
        else:
            print("not in teract")
            
    # Click add to cart button when selected all variant option
    # Wait for add to cart can access
    add_to_cart = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@data-atd='variant-add-to-cart-button']")))
    
    # Wait until can click add item to cart with click button many time
    while True:
        class_check = driver.find_element(By.XPATH, "//div[@data-atd='variant-modal-bottom-sheets']").get_attribute('class')
        
        if "pointer-events-none opacity-0 -z-10" in class_check:
            break
        else:
            add_to_cart.click()
            time.sleep(0.3)
    
    # If process finish close tab and sawp to original tab to open another product
    driver.close()
    driver.switch_to.window(tab_handle[0])
    
```

```python
# Access to shop cart

# Click to cart button to place order interface
cart_button = driver.find_element(By.XPATH, "//button[@data-atd='cart-button']")
cart_button.click()

# Select all product in cart
cart_checkbox = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@data-atd='cart-checkbox']")))
cart_checkbox.click()

# Then click to buy now and checkout
cart_buynow = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@data-atd='cart-buynow-button']")))
cart_buynow.click()

# Wait for can select shipping option
wait.until(EC.url_to_be(driver.current_url))
shipment_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@id='shipment-option-section']/div[@class='mt-4 select-none']/div")))
time.sleep(0.3)

# Check if shipment option doesn't any text then it have, Click it
while True:
    if shipment_option.text != "":
        break
    print("Can't")
shipment_option.click()


# Check payment and scroll to see element to perform any action
actions.scroll_by_amount(0,500).perform()
time.sleep(0.3)

# Select option to pay
select_payment = "credit/debit card"

try:    
    payment_title = driver.find_element(By.XPATH, "//div[@data-atd='payment-title-rlp']")
    payment_option = payment_title.find_elements(By.XPATH, "../div[@class='mb-3 select-none']/div")
    
    # Select the right payment
    for payment in payment_option:
        if select_payment in payment.text:
            payment.click()
            break
        
except: # Catch when select destination payment
    print("Don't pay now")

# Place order button
placeorder_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@data-atd='placeorder-button']")))
actions.scroll_to_element(placeorder_button)

placeorder_button.click()
```
